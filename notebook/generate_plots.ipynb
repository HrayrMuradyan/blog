```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
# Parameters for the distribution
mean_time = 4  
std_dev_time = 1  
num_students = 700  

# Generate random times following a normal distribution
completion_times = np.random.normal(mean_time, std_dev_time, num_students)

# Ensure no negative times (truncate at 0)
completion_times = np.clip(completion_times, 0, None)

mean_completion_time = round(np.mean(completion_times))

# Plot the distribution
plt.figure(figsize=(8, 5))
plt.hist(completion_times, bins=40, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Homework Completion Times', size=15)
plt.xlabel('Time to Complete Homework (hours)', size=13)
plt.ylabel('Number of Students', size=13)
plt.axvline(mean_completion_time, color='red', linestyle='dashed', linewidth=3, label='Mean')
plt.annotate(
    f'Average completion\ntime: {mean_completion_time} hours',
    xy=(mean_completion_time, 37),  # Point to the mean line
    xytext=(mean_completion_time + 2.4, 45),  # Position of the text
    arrowprops=dict(facecolor='black', arrowstyle='->', linewidth=3),
    fontsize=16,
    weight='bold',
    color='darkred',
    ha='center',
    va='center'
)
plt.savefig("../img/homework_completion_distribution.png")
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

np.random.seed(42)
# Parameters for the distribution
num_students = 700  

# Right-skewed (positive skew) data
right_skew_param = 10  # Positive value creates right skew
right_skew_data = skewnorm.rvs(a=right_skew_param, loc=2, scale=1, size=num_students)

# Left-skewed (negative skew) data
left_skew_param = -10  # Negative value creates left skew
left_skew_data = skewnorm.rvs(a=left_skew_param, loc=4, scale=1, size=num_students)

# Define common x-ticks
xticks = np.arange(0, 7, 1)

# Create subplots with two rows
fig, axes = plt.subplots(2, 1, figsize=(15, 8))

# Plot for right-skewed data
axes[0].hist(right_skew_data, bins=40, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].set_title('Right Skewed Homework Completion Times', size=22)
axes[0].set_xlabel('Time to Complete Homework (hours)', size=16)
axes[0].set_ylabel('Number of Students', size=16)
axes[0].set_xticks(range(0, 7))
right_mean = round(np.mean(right_skew_data))
axes[0].axvline(right_mean, color='red', linestyle='dashed', linewidth=4, label='Mean')
axes[0].annotate(
    f'Average completion\ntime: {right_mean} hours',
    xy=(right_mean, 37),  # Point to the mean line
    xytext=(right_mean + 1.8, 45),  # Position of the text
    arrowprops=dict(facecolor='black', arrowstyle='->', linewidth=4),
    fontsize=16,
    weight='bold',
    color='darkred',
    ha='center',
    va='center'
)

axes[0].tick_params(axis='x', labelsize=15)
axes[0].tick_params(axis='y', labelsize=15)

# Plot for left-skewed data
axes[1].hist(left_skew_data, bins=40, color='lightgreen', edgecolor='black', alpha=0.7)
axes[1].set_title('Left Skewed Homework Completion Times', size=22)
axes[1].set_xlabel('Time to Complete Homework (hours)', size=16)
axes[1].set_ylabel('Number of Students', size=16)
axes[1].set_xticks(xticks)
left_mean = round(np.mean(left_skew_data))
axes[1].axvline(left_mean, color='red', linestyle='dashed', linewidth=4, label='Mean')
axes[1].annotate(
    f'Average completion\ntime: {left_mean} hours',
    xy=(left_mean, 37),  # Point to the mean line
    xytext=(left_mean - 1.8, 45),  # Position of the text
    arrowprops=dict(facecolor='black', arrowstyle='->', linewidth=4),
    fontsize=16,
    weight='bold',
    color='darkred',
    ha='center',
    va='center'
)
axes[1].tick_params(axis='x', labelsize=15)
axes[1].tick_params(axis='y', labelsize=15)

plt.tight_layout()
plt.savefig("../img/skewed_homework_completion_distribution2.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)
# Parameters for the distribution
mean_time = 4  
std_dev_time = 1  
num_students = 700  

# Generate random times following a normal distribution
completion_times = np.random.normal(mean_time, std_dev_time, num_students)

# Ensure no negative times (truncate at 0)
completion_times = np.clip(completion_times, 0, None)

mean_completion_time = round(np.mean(completion_times))

# Generate x values for the normal curve
x = np.linspace(0, 8, 500)  # Values from 0 to 8 for the x-axis
pdf = norm.pdf(x, mean_time, std_dev_time)  # Normal PDF values

# Scale the PDF to match the histogram's scale
hist_values, bins, _ = plt.hist(completion_times, bins=40, color='skyblue', edgecolor='black', alpha=0.7)
scale_factor = num_students * (bins[1] - bins[0])  # Scale factor based on bin width and sample size
pdf_scaled = pdf * scale_factor

# Create the single plot
plt.figure(figsize=(10, 6))
plt.hist(completion_times, bins=40, color='skyblue', edgecolor='black', alpha=0.7, label='Histogram')
plt.plot(x, pdf_scaled, color='darkorange', lw=2, label='Normal Curve')
plt.axvline(mean_completion_time, color='red', linestyle='dashed', linewidth=3, label='Mean')
plt.annotate(
    f'Average completion\ntime: {mean_completion_time} hours',
    xy=(mean_completion_time, 37),  # Point to the mean line
    xytext=(mean_completion_time + 2.4, 45),  # Position of the text
    arrowprops=dict(facecolor='black', arrowstyle='->', linewidth=3),
    fontsize=16,
    weight='bold',
    color='darkred',
    ha='center',
    va='center'
)

# Add the PDF formula with a background box
pdf_formula = r"$f(x) = C \times \frac{1}{\sqrt{2\pi}} \times e^{-\frac{(x - 4)^2}{2}}$"
plt.annotate(
    pdf_formula,
    xy=(6, 30),  # Position for the formula
    xytext=(1.5, 35),
    fontsize=18,
    color='darkorange',
    weight='bold',
    ha='center',
    bbox=dict(boxstyle="round,pad=0.4", edgecolor='black', facecolor='whitesmoke', alpha=0.8)
)

plt.annotate(
    f'Probability Density Function',
    xy=(5.8, 30),
    xytext=(1.5, 43),
    fontsize=14,
    weight='bold',
    color='darkorange',
    ha='center',
    va='center'
)

# Add labels and title
plt.title('Distribution of Homework Completion Times with Probability Density Function', size=16)
plt.xlabel('Time to Complete Homework (hours)', size=14)
plt.ylabel('Number of Students', size=14)
plt.tight_layout()

# Save and show the plot
plt.savefig("../img/homework_completion_distribution_with_curve.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distributions
params = [
    {"mu": 0, "sigma": 1, "label": "=0, =1"},
    {"mu": 1, "sigma": 0.5, "label": "=1, =0.5"},
    {"mu": -2, "sigma": 1.5, "label": "=-2, =1.5"},
    {"mu": 2, "sigma": 1, "label": "=2, =1"},
    {"mu": -1, "sigma": 0.7, "label": "=-1, =0.7"},
]

# Generate x values
x = np.linspace(-5, 5, 500)

# Create the plot
plt.figure(figsize=(10, 6))

for param in params:
    y = (1 / (param["sigma"] * np.sqrt(2 * np.pi))) * \
        np.exp(-0.5 * ((x - param["mu"]) / param["sigma"])**2)
    plt.plot(x, y, label=param["label"])

# Add legend and labels
plt.title("Normal Distributions with Different Parameters", fontsize=16)
plt.xlabel("Variable", fontsize=14)
plt.ylabel("Probability Density", fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)
plt.savefig("../img/different_normal_distributions.png")
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the numbers on a dice
numbers = np.array([1, 2, 3, 4, 5, 6])

# The probability for each number on a fair die is 1/6
probabilities = np.array([1/6] * 6)

# Plotting
plt.figure(figsize=(6, 4))
bars = plt.bar(numbers, probabilities, color='skyblue')
plt.xlabel('Number on the Dice')
plt.ylabel('Probability')
plt.title('Probability of Rolling a Dice')
plt.ylim(0, 0.2)
plt.xticks(numbers)

# Adding the probability text on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.3f}', ha='center', va='bottom')

plt.savefig("../img/dice_probability.png")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Parameters
n_rolls = 5  # Initial number of rolls per sample
max_samples = 500  # Maximum number of samples

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Function to update the plot for the animation
def update(num):
    ax.clear()
    
    # Increase the number of rolls (sample size) for each iteration
    samples = np.random.randint(1, 7, size=(num, n_rolls))
    averages = samples.mean(axis=1)
    
    # Plot the histogram of the averages
    ax.hist(averages, bins=20, color='skyblue', edgecolor='black', alpha=0.8)
    ax.set_title(f'Histogram of Die Roll Averages (Samples: {num})')
    ax.set_xlabel('Average of Rolls')
    ax.set_ylabel('Frequency')
    ax.set_xlim(1, 6)
    scale = len(samples) // 100
    ax.set_ylim(0, 20*(scale+1))
    ax.axvline(3.5, color='red', linestyle='--', linewidth=2, label="Expected Average (3.5)")
    # Add text to show the current number of samples
    ax.text(4.5, 12, f'Samples: {num}', fontsize=12, color='black')
    ax.legend()
    
# Create the animation
ani = animation.FuncAnimation(fig, update, frames=range(1, max_samples + 1), interval=200)

# Save the animation as a GIF
ani.save('../img/clt_die_rolls_increasing_samples.gif', writer='imagemagick')
```
